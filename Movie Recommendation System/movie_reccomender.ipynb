```python
ratings_with_name = ratings.merge(movies,on='movieId')
ratings_with_name
```

```python
ratings_with_name.info()
```

```python
x = ratings_with_name.groupby('userId').count()['rating']>=250
mature = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['userId'].isin(mature)]
y = filtered_rating.groupby('title').count()['rating']>=25
famous_movies = y[y].index
```

```python
final_ratings = filtered_rating[ filtered_rating['title'].isin(famous_movies)]
pt = final_ratings.pivot_table(index = 'title', columns = 'userId',values = 'rating')
```

```python
pt.fillna(0,inplace = True )
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(movie_title):
    
    index = np.where(pt.index == movie_title)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse = True)[1:6]

    for i in similar_items:
        print(pt.index[i[0]])
        
```
